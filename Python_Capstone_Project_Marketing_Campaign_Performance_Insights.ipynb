 **Exploratory Data Analysis (EDA) and Visualization**
 **Campaign Performance **

```python
#Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Acquisition_Cost',y='ROI',hue='Acquisition_Cost')
plt.title('The distribution of relationship between Acquisition_Cost and ROI')
plt.show()
```

```python
# Conversion_Rate=df['Conversion_Rate'].mean()
# Conversion_Rate
```

```python
#Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Channel_Used',y='Conversion_Rate',hue='Campaign_Type')
plt.title('The distribution of average Conversion_Rate for different Channel_Used')
plt.xticks(rotation=45)
plt.show()
```

```python
#Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Engagement_Score',y='Campaign_Type',hue='Campaign_Type')
plt.title('The distribution of Engagement_Score across different Campaign_Type')
plt.show()
```

```python
# Analyze the average ROI
# ROI=df['ROI'].mean()
# ROI
```

```python
# Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Company',y='ROI',hue='Campaign_Type')
plt.title('The distribution of average ROI by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
#Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.
numerical=df[['Engagement_Score','Conversion_Rate']].corr()
numerical
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(numerical,annot=True)
plt.figure(figsize=(10,6))
plt.show()
```

```python
#Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Language',y='Conversion_Rate',hue='Campaign_Type')
plt.title('The Distribution of Average Conversion_Rate by Language')
plt.xticks(rotation=45)
plt.show()
```

```python
#Show the distribution of total ROI across different Channels_Used using a pie chart.
groupby=df.groupby('Channel_Used').count().reset_index()
groupby
y=groupby['ROI']
labels=groupby['Channel_Used']
plt.figure(figsize=(10,6))
plt.pie(y,labels=labels,autopct='%1.2f%%',shadow=True,startangle=90,explode=(0,0.05,0,0,0,0))
plt.title('The distribution of total ROI across different Channels_Used')
plt.show()
```

 **Time-Base Analysis**

```python
#Plot the distribution of Duration using a histogram.
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Duration',hue='Duration')
plt.title('The distribution of Duration')
plt.xticks(rotation=45)
plt.show()
```

```python
#Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.
import matplotlib.dates as mdates
plt.figure(figsize=(10, 6))
sns.lineplot(data=df,x='Date',y='Conversion_Rate',hue='Company',errorbar=None)
Date = pd.date_range(start="01-01-2023", periods=100, freq='D')
step = 7
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
# plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()






```

```python
#Examine the trend of Engagement_Score over Date with a line chart.

plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x='Date', y='Engagement_Score',hue_order='Engagement_Score',errorbar=None)
Date = pd.date_range(start="01-01-2023", periods=100, freq='D')
step = 7
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.title("Trend of Engagement Score Over Date")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.show()


```

 **Geographic Analysis :**

```python
#Determine which location has the highest Acquisition_Cost using a bar chart
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Location',y='Acquisition_Cost',hue='Location')
plt.title('location has the highest Acquisition_Cost')
plt.xticks(rotation=45)
plt.show()
```
